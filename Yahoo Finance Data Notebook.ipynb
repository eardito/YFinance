```python
import pandas as pd
from pandas_datareader import data

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tsla = data.DataReader(name = "TSLA", data_source = "yahoo", start = "2019-11-01", end = "2019-12-31")
tsla.head(5)
```

```python
pfe = data.DataReader(name = 'PFE', data_source = 'yahoo', start = '2019-11-01', end='2019-12-31' )
pfe.head(5)
```

```python
tsla['Adj Close'].plot()
```

```python
plt.plot(tsla.index, tsla['Adj Close'])
plt.xticks(rotation = 60     )
plt.xlabel('Month')
plt.ylabel('Stock Price')
plt.title('TSLA')
```

```python
fig = plt.figure(figsize = (12,12))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(tsla.index, tsla['Adj Close'])
ax2.plot(pfe.index, pfe['Adj Close'])
ax1.set_title('TSLA Chart')
ax2.set_title('PFE Chart')
```

```python
plt.plot(tsla.index, tsla['Adj Close'])
plt.plot(pfe.index, pfe['Adj Close'])
```
